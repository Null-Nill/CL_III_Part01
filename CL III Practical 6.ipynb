## Problem Statement - Implementation of Clonal selection algorithm using Python.

```python
import numpy as np
import random
```

```python
# Objective function (Affinity)
# Maximum value at x=0, y=0
def affinity_function(antibody):
    x = antibody[0]
    y = antibody[1]
    return -(x**2 + y**2) + 100
```

```python
# Step 1: Initialize population
def initialize_population(size, lower, upper):
    population = []
    for i in range(size):
        x = random.uniform(lower, upper)
        y = random.uniform(lower, upper)
        population.append([x, y])
    return population
```

```python
# Step 2: Evaluate population
def evaluate_population(population):
    affinities = []
    for ind in population:
        affinities.append(affinity_function(ind))
    return affinities
```

```python
# Step 3: Clone (more clones for better solutions)
def clone_population(population, affinities, clone_factor):
    clones = []
    
    for i in range(len(population)):
        # better affinity → more clones
        num_clones = int(clone_factor / (i + 1))
        
        for j in range(num_clones):
            clones.append(population[i][:])  # copy
    
    return clones
```

```python
# Step 4: Mutation
def mutate(clones, mutation_rate, lower, upper):
    for ind in clones:
        if random.random() < mutation_rate:
            # small change
            ind[0] += random.uniform(-0.5, 0.5)
            ind[1] += random.uniform(-0.5, 0.5)
            
            # keep values in range
            ind[0] = max(lower, min(upper, ind[0]))
            ind[1] = max(lower, min(upper, ind[1]))
    
    return clones
```

```python
# Step 5: Selection (best individuals survive)
def select_best(population, size):
    population.sort(key=lambda x: affinity_function(x), reverse=True)
    return population[:size]
```

```python
# Main Program
# -----------------------------------

POP_SIZE = 10
GENERATIONS = 20
LOWER = -5
UPPER = 5
CLONE_FACTOR = 10
MUTATION_RATE = 0.3

# Initialize
population = initialize_population(POP_SIZE, LOWER, UPPER)

print("Starting Clonal Selection Algorithm...\n")
```

```python
# Run algorithm
for gen in range(GENERATIONS):
    print("Generation", gen+1)
    
    # Evaluate
    affinities = evaluate_population(population)
    
    # Sort population (best first)
    combined = list(zip(population, affinities))
    combined.sort(key=lambda x: x[1], reverse=True)
    population = [item[0] for item in combined]
    
    # Clone
    clones = clone_population(population, affinities, CLONE_FACTOR)
    
    # Mutation
    clones = mutate(clones, MUTATION_RATE, LOWER, UPPER)
    
    # Combine and select best
    population = population + clones
    population = select_best(population, POP_SIZE)
    
    # Print best solution of this generation
    best = population[0]
    best_aff = affinity_function(best)
    print("Best:", best, "Affinity:", best_aff)
    print()
```

```python
# Final result
best = population[0]
print("Final Best Solution:", best)
print("Final Affinity:", affinity_function(best))
```
