## Problem Statement - Optimization of genetic algorithm parameter in hybrid genetic algorithm-neural network modelling: Application to spray drying of coconut milk.

```python
import random

#pip install deap
from deap import base, creator, tools, algorithms
```

```python
# 1. Dummy Dataset (simulating problem)
# -------------------------------
X = [0.1, 0.2, 0.3, 0.4]
y = [0.2, 0.4, 0.6, 0.8]
```

```python
# 2. Simple Neural Network (very basic)
# -------------------------------
def simple_nn(x, neurons, layers):
    # simple logic just for simulation
    output = x
    for _ in range(layers):
        output = output * neurons * 0.1
    return output
```

```python
# 3. Fitness Function (MSE error)
# -------------------------------
def evaluate(individual):
    neurons, layers = individual
    error = 0
    
    for xi, yi in zip(X, y):
        pred = simple_nn(xi, neurons, layers)
        error += (pred - yi) ** 2
    
    return error
```

```python
# 4. Genetic Algorithm Parameters
# -------------------------------
POP_SIZE = 6
GENERATIONS = 5

# Create initial population
population = []
for _ in range(POP_SIZE):
    neurons = random.randint(1, 10)
    layers = random.randint(1, 5)
    population.append([neurons, layers])
```

```python
# 5. Genetic Operations
# -------------------------------
def selection(pop):
    pop.sort(key=lambda ind: evaluate(ind))
    return pop[:3]  # select best 3

def crossover(p1, p2):
    return [p1[0], p2[1]]

def mutation(ind):
    if random.random() < 0.3:
        ind[0] = random.randint(1, 10)
    if random.random() < 0.3:
        ind[1] = random.randint(1, 5)
    return ind
```

```python
# 6. Run Genetic Algorithm
# -------------------------------
for gen in range(GENERATIONS):
    print(f"\nGeneration {gen+1}")
    
    # Selection
    selected = selection(population)
    
    # Create new population
    new_population = selected.copy()
    
    while len(new_population) < POP_SIZE:
        p1 = random.choice(selected)
        p2 = random.choice(selected)
        
        child = crossover(p1, p2)
        child = mutation(child)
        
        new_population.append(child)
    
    population = new_population
    
    # Print population
    for ind in population:
        print("Individual:", ind, "Error:", evaluate(ind))
```

```python
# 7. Best Solution
# -------------------------------
best = min(population, key=lambda ind: evaluate(ind))

print("\nBest Parameters Found:")
print("Neurons =", best[0], "Layers =", best[1])
```
